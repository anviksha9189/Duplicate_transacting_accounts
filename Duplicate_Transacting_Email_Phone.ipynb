```python
def check_common_element(a,b):
    for i in a: 
        if i == '??':
            continue
        if i in b:
            return True
    return False
```

```python
df2['Is_Duplicate_Transacting'] = 'No'
```

```python
for i in df1.index:
    for j in df2.index:
        if fuzz.ratio(df1['business_name'][i].lower(), df2['business_name'][j].lower()) > 85 and df1['business_name'][i].lower() != '****' and df2['business_name'][j].lower() != '****' and len(df2['business_name'][j]) > 2:
            if check_common_element(df1['user_email'][i], df2['user_email'][j]):
                df2['Is_Duplicate_Transacting'][j] = 'Yes'
            if check_common_element(df1['user_mobile_number'][i], df2['user_mobile_number'][j]) and check_common_element(df1['user_mobile_country_code'][i], df2['user_mobile_country_code'][j]):
                df2['Is_Duplicate_Transacting'][j] = 'Yes'
    stri = str(i) + 'EmPhDupRem' + '.csv'
    if i%100 == 0:
        df2.to_csv(stri)
    print(i)
```
